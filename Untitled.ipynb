```python
class MLAgent():
    def __init__(self, action_size, ML_class, parameters = {"n_jobs" : -1}, seed = None):
        """
        Initalize a Machine learning agent.

        Arguments
        ---------
        action_size: int
            the size of the action.
        ML_class: ModelClass
            the class of the machine learning model used for this agent.
        parameters: dict()
            the parameters for the model.
        seed: int
            the rng parameter for numpy.
        """
        self._rng = np.random.RandomState(seed = seed)

        self.action_size = action_size
        self.ML_class = ML_class
        self.parameters  = parameters

    def act(self, context):
        """
        Choose an action.

        Arguments
        ---------
        context: list()
            the context.
        
        Return
        ---------
        action: np.ndarray
            the action.
        """
        X = np.zeros((len(context), len(context[0][2])))

        for i, (_, _, variables) in enumerate(context):
            X[i, :] = variables

        predictions = self.model.predict(X)

        idx_action  = np.argsort(predictions)[-self.action_size:][::-1].astype(int)

        action = [context[idx][1] for idx in idx_action]

        return action


    def update(self, context, action, reward):
        """
        Update the agent. Not relevent for this agent because the online learning has not been implemented yet.
        """
        pass


    def train_agent(self, contexts, actions, rewards, clicks):
        """
        Train the agent using an history.

        Arguments
        ---------
        contexts: list()
            list of the history contexts.
        actions: np.ndarray
            the history of actions.
        rewards: np.ndarray
            the history of rewards.
        clicks: np.ndarray
            the history of clicks.
        """

        steps          = len(contexts)
        variables_size = len(contexts[0][0][2])

        X = np.zeros((steps * self.action_size, variables_size))
        y = np.zeros((steps * self.action_size))
        
        for step in range(steps):
            for i in range(self.action_size):
                X[step * self.action_size + i] = contexts[step][actions[step].astype(int)[i]][2]
                y[step * self.action_size + i] = rewards[step, i]

        self.model = self.ML_class(**self.parameters)
        self.model.fit(X, y)
```

```python
def run_exp_hist(random_agent, agent, env, nb_steps_history, nb_steps, action_size, env_seed):
    """
    Run an experiment for an agent with an history made by a random agent.

    Arguments
    ---------
    random_agent: RandomAgent
        the random agent.
    agent: Agent
        the agent.
    env: Environment
        the environment.
    nb_steps_history: int
        number of steps for the history.
    nb_steps: int
        number of steps for the experiment.
    action_size: int
        the action size.
    env_seed: int
        seed for the rng of the environment.

    Return
    -------
    dict
        the results of the experiment.
    """
    rewards = np.zeros((nb_steps, action_size))
    regrets = np.zeros((nb_steps, action_size))
    actions = np.zeros((nb_steps, action_size))
    clicks  = np.zeros((nb_steps, action_size))

    contexts_hist = []
    rewards_hist = np.zeros((nb_steps_history, action_size))
    actions_hist = np.zeros((nb_steps_history, action_size))
    clicks_hist  = np.zeros((nb_steps_history, action_size))

    context = env.reset(env_seed)

    for i in range(nb_steps_history):
        # Select action from agent policy.
        action = random_agent.act(context)
        
        # Play action in the environment and get reward.
        next_context, reward, click, best_reward = env.step(action)
        
        # Update history
        rewards_hist[i] = reward
        actions_hist[i] = action
        clicks_hist[i]  = click
        contexts_hist   += [context]
        
        # Update agent.
        random_agent.update(context, action, reward)
        context = next_context
        
    agent.train_agent(contexts_hist, actions_hist, rewards_hist, clicks_hist)
    
    for i in range(nb_steps):
        # Select action from agent policy.
        action = agent.act(context)
        
        # Play action in the environment and get reward.
        next_context, reward, click, best_reward  = env.step(action)

        
        # Update agent.
        agent.update(context, action, reward)
        context = next_context
        
        # Save history.
        rewards[i] = reward
        actions[i] = action
        regrets[i] = best_reward - reward
        clicks[i]  = click
        if i in range(nb_steps -5 , nb_steps) : 
            print("best reward is : " , best_reward)
            print("reward is : " , reward)
            print("regrets is : " , regrets[i])

    reward = rewards[:, 0].sum()
    regret = np.sum(regrets, axis = 0)

    return {'reward': reward, 
            'regret': regret,
            'rewards': rewards,
            'regrets': regrets,
            'actions': actions,
            'cum_rewards': np.cumsum(rewards, axis = 0), 
            'cum_regrets': np.cumsum(regrets, axis = 0),
            }
```
